```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    'Employee': ['John', 'Sarah', 'Mike', 'Emma', 'Tom'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [60000, 50000, 75000, 65000, 48000],
    'Experience (Years)': [5, 3, 7, 4, 2],
    'Performance Rating (Out of 5)': [4.5, 3.8, 4.9, 4.2, 3.5]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
avg_salary_by_dept=df.groupby('Department')['Salary'].mean()
avg_salary_by_dept
```

```python
high_performers=df[df['Performance Rating (Out of 5)']>=4.0]
high_performers
```

```python
def calculate_increment(rating):
  if rating>=4.0:
    return 0.10
  else:
    return 0.05
df['Increment']=df['Performance Rating (Out of 5)'].apply(calculate_increment)
df
```

```python
df['salary_after_increment']=df['Salary']+(df['Salary']*df['Increment'])
df
```

```python
sorted_df = df.sort_values(by='salary_after_increment', ascending=False)
print(sorted_df)
```

```python
cdata={
    'City':['New York','Los Angeles','Chicago','Houston','Miami'],
    'Total Cases':[100000,85000,65000,70000,55000],
    'Recovered':[75000,70000,50000,60000,45000],
    'Deaths':[5000,3500,2500,2000,1500]
}
```

```python
df=pd.DataFrame(cdata)
df
```

```python
df['recovery_rate']=(df['Recovered']/df['Total Cases'])*100
df
```

```python
df['death_rate']=(df['Deaths']/df['Total Cases'])*100
df
```

```python
death_rate_by_city=df[df['death_rate']>=0.05]
death_rate_by_city
```

```python
sorted_df = df.sort_values(by='recovery_rate', ascending=False)
print(sorted_df)
```

```python
total_cases=df['Total Cases'].sum()
total_deaths=df['Deaths'].sum()
print("Total Cases:",total_cases)
print("Total Deaths:",total_deaths)
```
