```python
import pandas as pd
```

encoding

```python
col_data = {'color':['Red','Green', 'Blue','Yellow','Black','White']}
df_col = pd.DataFrame(col_data)
```

perform one hot encoding

```python
one_hot_encoded = pd.get_dummies(df_col['color'],prefix = 'Color')
```

```python
one_hot_encoded = one_hot_encoded.astype(int)
```

```python
print(one_hot_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df_col['Color'] = label_encoder.fit_transform(df_col['color'])
```

```python
print("\nDataFrame with Color_Label column:\n", df_col)
```
