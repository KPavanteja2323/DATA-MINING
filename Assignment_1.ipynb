```python
import pandas as pd

# Load the dataset
file_path = "/content/employee_data.csv"  # Ensure the file is uploaded to Google Colab
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

```python
2.# Check for missing values
df.isnull().sum()
```

```python
# Impute missing values
df['Age'].fillna(df['Age'].median(), inplace=True)  # Median for Age
df['Salary'].fillna(df['Salary'].mean(), inplace=True)  # Mean for Salary
df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode()[0], inplace=True)  # Mode for Job Satisfaction

# Check again to confirm missing values are handled
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Apply Min-Max Scaling
minmax_scaler = MinMaxScaler()
df[['Age_Scaled', 'Salary_Scaled']] = minmax_scaler.fit_transform(df[['Age', 'Salary']])

# Display first few rows after scaling
df[['Age', 'Age_Scaled', 'Salary', 'Salary_Scaled']].head()
```

```python
df['Age_Manual_Scaled'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Salary_Manual_Scaled'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

# Display results
df[['Age', 'Age_Manual_Scaled', 'Salary', 'Salary_Manual_Scaled']].head()
```

```python
from sklearn.preprocessing import StandardScaler

# Apply Standardization
standard_scaler = StandardScaler()
df[['Job_Satisfaction_Std', 'Work_Hours_Std']] = standard_scaler.fit_transform(df[['Job_Satisfaction', 'Work_Hours_Per_Week']])

# Display first few rows
df[['Job_Satisfaction', 'Job_Satisfaction_Std', 'Work_Hours_Per_Week', 'Work_Hours_Std']].head()
```

```python
df['Job_Satisfaction_Manual_Std'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()
df['Work_Hours_Manual_Std'] = (df['Work_Hours_Per_Week'] - df['Work_Hours_Per_Week'].mean()) / df['Work_Hours_Per_Week'].std()

# Display results
df[['Job_Satisfaction', 'Job_Satisfaction_Manual_Std', 'Work_Hours_Per_Week', 'Work_Hours_Manual_Std']].head()
```
